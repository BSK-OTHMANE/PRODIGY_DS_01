```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("task1.csv")
df
```

```python
# Step 1: Remove unnamed column(s)
df = df.loc[:, ~df.columns.str.contains('Unnamed: 68')]
columns_to_drop = ['Country Code', 'Indicator Name', 'Indicator Code']
df = df.drop(columns_to_drop, axis=1)
arabworld_df = df[df['Country Name'] == 'Arab World']
arabworld_df
```

```python
years = arabworld_df.columns[1:]  # Assuming the first four columns are metadata
population = arabworld_df.iloc[0, 1:].values
```

```python
plt.figure(figsize=(20 ,10))
plt.bar(years, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Aruba Over the Years')
plt.xticks(rotation=90)
plt.show
```
